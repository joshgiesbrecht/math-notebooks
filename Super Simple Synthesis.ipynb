## Audio Synthesis Using Trig Functions!

#### Super simple synthesis 

First we do some setting up of the programming environment, telling it what code libaries to refer to for things like sin() and plot(): 

```python
%matplotlib inline
from matplotlib.pyplot import *
from numpy import *
from IPython.display import Audio
```

Then we set up a few variables we'll need:

```python
Fs = 44100 ## sample rate (44100 is usual for digital audio)
Len = 3    ## length of sound in seconds
t = linspace(0, Len, Fs*Len)   ## time variable
```

And now, the good stuff: creating the actual audio sample and giving it a Play button.

```python
sine_sample = sin(2*pi*440*t)
Audio(data=sine_sample, rate=Fs)
```

Not so bad! With under a dozen lines of code we can now create our own samples.

The most important line for our purposes is: `sine_sample = sin(2*pi*440*t)` above.
Try making changes to what you're sending to the sin() function. What does the 440 represent?

Then try making something more complicated. Perhaps two interacting sine waves with slightly different frequencies?
`sine_sample = sin(2*pi*440*t) + sin(2*pi*438*t)`

We can also plot what that sine wave sample looks like:

```python
plot(sine_sample[0:500])
```

Note that we're only plotting up to an x value of 500 so that the wave is visible; if we fit the entire three seconds worth of audio into one graph, it would be too full to see what's happening.
